Library Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import klib
```

**Deskripsi Dataset:**

* **Survival**: 0 (tidak selamat) atau 1 (selamat)
* **Pclass**: kelas tiket
* **Sex**: laki-laki (*male*) atau perempuan (*female*)
* **Age**: umur penumpang
* **Sibsp**: jumlah saudara yang ikut naik kapal, misalnya adik, kakak, dan istri/ suami.
* **Parch**: jumlah anggota relasi yang umurnya jauh berbeda, yang ikut naik kapal, misalnya orang tua dan anak-anak.
* **Ticket**: nomor tiket
* **Fare**: biaya tiket
* **Cabin**: nomor kabin
* **Embarked**: tempat awal pelabuhan dengan 3 variasi *value*, yaitu Cherbourg (C), Queenstown (Q), dan Southampton (S). 

Deskripsi Dataset dilihat di: https://www.kaggle.com/competitions/titanic/data.

## Two Variables

### Categorical Variables

**Stacked Bar Chart: Comparison & Composition**

```python
stacked_bar_chart_width = 0.9
survived_0_color = 'red'
survived_1_color = 'green'

figure = plt.figure()
figure.set_figwidth(15)
figure.set_figheight(15)

stacked_bar_fields = categorical_variables.copy()
stacked_bar_fields.append(survived_field)

for index, field in enumerate(stacked_bar_fields):
    
    if field == survived_field:
        continue

    plt.subplot(4, 3, index)

    sum_survived_0 = titanic_df[titanic_df[survived_field] == 0].groupby(field)[survived_field].count()
    sum_survived_1 = titanic_df[titanic_df[survived_field] == 1].groupby(field)[survived_field].count()

    unique_value_0 = list(sum_survived_0.index)[:5]
    unique_value_1 = list(sum_survived_1.index)[:5]

    sum_survived_0 = sum_survived_0.values.flatten()[:5]
    sum_survived_1 = sum_survived_1.values.flatten()[:5]

    plt.title(field)

    plt.bar(unique_value_0, sum_survived_0, color = survived_0_color, width=stacked_bar_chart_width)
    plt.bar(unique_value_1, sum_survived_1, color = survived_1_color, width=stacked_bar_chart_width, bottom=sum_survived_0)

plt.show()
```

**Insight**:

* **PClass:**
  * Persentase penumpang yang mengambil tiket kelas 1 lebih banyak yang selamat.
  * Jumlah penumpang yang membeli tiket kelas 3 lebih banyak. 
  * Jumlah penumpang yang membeli tiket kelas 1 sama banyak dengan jumlah orang yang membeli tiket kelas 2.

* **SibSp:**
  * Persentase penumpang dengan jumlah saudaranya ikut di bawah 3 lebih banyak yang selamat.
  * Jumlah penumpang yang paling banyak adalah penumpang yang tidak membawa saudara.
  * Jumlah penumpang yang membawa 1 saudara adalah terbanyak kedua.
  * Jarang ditemukan penumpang yang membawa 2/ lebih saudara pada kapal.

* **Parch:**
  * Persentase penumpang dengan nilai parch 1 atau 2 lebih banyak yang selamat.
  * Jumlah terbanyak adalah penumpang dengan nilai parch 0.
  * Jumlah penumpang dengan nilai parch 1 dan 2 hampir setara.
  * Jarang ditemukan penumpang dengan nilai parch lebih dari 2.
  
* **Sex:**
  * Persentase penumpang perempuan yang selamat lebih besar daripada laki-laki.
  * Jumlah penumpang laki-laki lebih banyak daripada perempuan.

* **Embarked:**
  * Persentase penumpang yang berangkat dari Cherbourg lebih banyak yang selamat daripada persentase pada Queenstown dan Southampton.
  * Jumlah orang yang memulai perjalanan dari Southampton lebih banyak daripada starting point lainnya.

## Multi Variable

### Categorical Variables 

**Multiple Bar Chart - Distribution**

```python
stacked_bar_chart_width = 0.3
survived_0_color = 'red'
survived_1_color = 'green'

stacked_bar_numerical_fields = numerical_variables.copy()
stacked_bar_categorical_fields = categorical_variables.copy()

# stacked_bar_categorical_fields.remove(parch_field)
# stacked_bar_categorical_fields.remove(pclass_field)

for numerical_index, categorical_field in enumerate(stacked_bar_categorical_fields):
    if (survived_field == categorical_field):
        continue

    figure = plt.figure()
    figure.set_figwidth(10)
    figure.set_figheight(5)

    for numerical_index, numerical_field in enumerate(stacked_bar_numerical_fields):
        
        plt.subplot(1, 2, numerical_index + 1)

        sum_survived_0 = titanic_df[titanic_df[survived_field] == 0].groupby(categorical_field)[numerical_field].mean()
        sum_survived_1 = titanic_df[titanic_df[survived_field] == 1].groupby(categorical_field)[numerical_field].mean()

        unique_value_0 = list(sum_survived_0.index)[:5]
        unique_value_1 = list(sum_survived_1.index)[:5]

        sum_survived_0 = sum_survived_0.values.flatten()[:5]
        sum_survived_1 = sum_survived_1.values.flatten()[:5]

        plt.title(f"{categorical_field} - {numerical_field} Mean")

        x_axis = np.arange(len(unique_value_0))

        plt.xticks(x_axis, unique_value_0)
        plt.xlabel(categorical_field)
        plt.ylabel(f"{numerical_field} Mean")

        plt.bar(x_axis - 0.2, sum_survived_0, width=stacked_bar_chart_width, label = 'Survived')
        plt.bar(x_axis + 0.2, sum_survived_1, width=stacked_bar_chart_width, label = 'Died')

        plt.legend()
    plt.show()
```

**Insight:**

* **PClass:**
  * Ketiga kelas pada tiket memiliki jumlah penumpang yang selamat lebih banyak daripada jumlah penumpang yang tidak selamat.
  * Penumpang yang lebih tua cenderung membeli tiket kelas 1 dibandingkan dengan tiket kelas 2 dan 3.
  * Harga tiket kelas 1 lebih mahal dibandingkan dengan tiket kelas lainnya.
  * Banyak yang tidak selamat dengan harga tiket yang mahal.

* **SibSp:**
  * Umur penumpang yang lebih tua dan membawa *sibling* dengan jumlah dibawah 3 memiliki nilai survived yang lebih tinggi.
  * Untuk penumpang yang membawa *sibling* dengan jumlah dibawah 4 dan membeli tiket yang mahal memiliki nilai died yang lebih tinggi.

* **Parch:**
  * Persentase penumpang yang tidak selamat dan selamat pada distribusi parch hampir sama.
  * Penumpang dengan jumlah parch 2 memiliki rata-rata yang paling muda.
  * Rata-rata harga tiket yang lebih mahal untuk penumpang dengan jumlah parch 3 ke atas memiliki nilai survived yang tinggi.

* **Sex:**
  * Secara rata-rata, penumpang laki-laki yang naik kapal memiliki umur yang lebih tua daripada umur perempuan.
  * Tiket yang dibeli oleh perempuan cenderung lebih mahal daripada tiket laki-laki.
  * Banyak penumpang laki-laki dan perempuan yang tidak selamat dengan harga rata-rata tiket yang mahal.

* **Embarked:**
  * Penumpang yang lebih tua memiliki nilai survived yang lebih tinggi dari semua starting point yang ada (Cherbourg, Queenstown, dan Southampton).
  * Penumpang dengan tiket yang mahal dan *starting point* pada Cherburg dan Southampton lebih banyak yang tidak selamat dibandingkan dengan tiket yang murah.
  * Penumpang di Cherburg banyak yang membeli tiket yang mahal.

```python
correlation_matrix_fields = numerical_variables.copy()
correlation_matrix_fields.append(sib_sp_field)
correlation_matrix_fields.append(parch_field)

numerical_correlation_matrix = titanic_df[correlation_matrix_fields].corr()

plt.figure(figsize = (5, 5))
sns.heatmap(numerical_correlation_matrix, annot=True)
plt.show()
```
